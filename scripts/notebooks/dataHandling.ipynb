# Useful functions to help filter the dataframe

```python
def filter( df : pd.DataFrame, label : str, value,
           returnBool : bool = False                ) -> pd.DataFrame | np.ndarray:
    # VERIFIED
    """
        Takes in a dataframe, a label and the accepted values. 
        
        Warning : the type of the df[label] and value must match

        returns
            - if returnBool = False : the dataframe filtered -> pd.DataFrame
            - if returnBool = True  : the boolean array to filter the dataframe manually
    """

    # we are using the logical operator _in_, thus we make the value a list of values if it wasn't
    if not isinstance(value, list) :
         value = [value]
    
    # Create a boolean array used to filter
    boolArray : np.ndarray = np.array(
        [item in value for item in df[label]],
        dtype = bool)
    
    if returnBool : return boolArray

    return df[boolArray]
```

```python
def getPerYear(df : pd.DataFrame, label : str = "bert_genre", func = np.mean, filters : list[dict] = []) -> tuple[list, list]:
    # VERIFIED
    """
        takes a dataframe (non filtered) and read the table given the column label (label). it applies the 
        function func (usually np.means).
        It is possible to apply a filter thanks to the function filter.
        the _filters_ parameter is a list of dictionnaries {label, value} to apply to the dataframe

        returns the data retrieved and the years (dataRetrieved, years)
    """
    
    # Prepare output
    out = []
    years = list(set(df['annee']))  # select the years

    # filter data
    for f in filters:
        df = filter(df, f['label'], f['value'])
    
    # Retrieve data
    for y in years : 
        out.append(
            func(
                filter(df, 'annee',y)[label]
            ))
        
    return out, years
```

# Save dataframe to plot later

```python
def saveDataFrameForPlotting(df, xlabel : str, ylabel : str,
                             **kwargs) -> None:
    """
        This function is meant to be the bridge between the filtered dataframe and the actual plotting.
        In order to be as free as possible, the function will take kwargs to match the plotly arguments
        (color, line_dash, symbol, hover_name, hover_data, facet_row, facet_col, error_x, error_y, error_x_minus, error_y_minus, animation_frame, animation_group, ...) 
        # UPGRADE the list of accepted arguments
    """

    # UPGRADE Modif fonctions pour les rendre modulaires
    #NOTE tryout
    df = df.copy()

    # Prepare metadata ouput : 
    # the metadata dictionnary will hold the parameters and the columns to refer to 
    metadata : dict[str:str] = {}
    if xlabel in df.columns: metadata['xlabel'] = xlabel
    else : raise KeyError(f'{xlabel} is not a column in the dataframe')
    
    if (ylabel in df.columns) & (ylabel != xlabel): metadata['ylabel'] = ylabel
    else : 
        if not (ylabel in df.columns)   : raise KeyError(f'{ylabel} is not a column in the dataframe')
        if not (ylabel != xlabel)       : raise KeyError(f'{ylabel} is the same as xlabel ({xlabel})')

    # loop through kwargs and find to find the plotly arguments : 
    PLOTLYARGS = ['color', 'line_dash', 'symbol', 'hover_name', 'hover_data', 'facet_row', 'facet_col', 'error_x', 'error_y', 'error_x_minus', 'error_y_minus', 'animation_frame', 'animation_group']
    for key in kwargs : 
        if (key in PLOTLYARGS)&(key != xlabel)&(key != ylabel):     #NOTE Might be unnecessary 
            # kwargs[key] is either a column name or a vector, the size of the plot that will be added to the dataframe
            
            if isinstance(kwargs[key],str):             # is a column name
                # Only adds information to the metadata file
                try:
                    assert(kwargs[key] in df.columns)   # is a column of the dataframe
                    metadata[key] = kwargs[key]
                except:
                    print(f'WARNING : The kwargs[{key}] = {kwargs[key]} is not a column of the data frame. Ignored.')
            else :
                # Add the data to the dataframe and save the information in the metadata file

                # Verify the length of the serie
                try :       assert(len(kwargs[key]) == len(df)) # make sure the length is correct
                except :    print(f'WARNING : kwargs[{key}] is not of length {len(df)} ({len(kwargs[key])}). Ignored.'); pass
                
                # try to add it in the dataframe
                try :
                    df.loc[:,key] = pd.DataFrame({
                                        'temp' : kwargs[key]
                                        }).set_index(df.index).loc[:,'temp']
                    metadata[key] = key
                except:     print(f'WARNING : kwargs[{key}] could not be added to the dataframe. Ignored'); pass

            

    # Saving zone
    #UPGRADE for now the files are overwritten, think of a way to move the document we are about to delete somewhere else
    if "filename" in kwargs :
        filename : str = kwargs['filename']
    else : 
        filename : str = 'someCSV'
    
    # We are going to remove the ".csv" to switch from csv to json files
    if (filename[-4:] == ".csv") : filename = filename[:-4]

    # Add the date
    dateSTR : str = str(date.today())
    filename = dateSTR + "-" + filename

    # save document 
    df.to_csv(SAVEPATH + filename + ".csv")

    # save metadata in json
    with open(SAVEPATH + filename + ".json", "w") as outfile:
        json.dump(metadata, outfile)
    print("Done")

    #TODELETE 
    print(metadata)
    
```
